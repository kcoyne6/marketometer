```python
import pandas as pd
from sqlalchemy import create_engine

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
file = "final_stock_data.csv"
stock_sql = pd.read_csv(file)

stock_sql.head()
```

```python
stock_sql=stock_sql.rename(columns = {'Adj Close':'Adj_Close'})
del stock_sql['Unnamed: 0']
stock_sql.head()
```

```python
engine = create_engine("sqlite:///final_stock_data.sqlite")
conn = engine.connect()
```

```python
stock_sql.to_sql(name='final_stock_data', con=engine, if_exists='replace', index=False)
# # stock_sql.to_sql('stock_data', engine, if_exists = 'replace')
```

```python
pd.read_sql_query('select * from final_stock_data', con=engine).head()
```
